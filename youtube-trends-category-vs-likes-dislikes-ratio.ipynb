```python

#Checking basic information about the canadian dataset
print("\nBasic information about the dataset")
print(CAvideos.info())

#checking satistics for numeric columns
print("\nSummary Statistics for the numeric columns")
print(CAvideos.describe())

#checking for missing values
print("\nMissing values:")
print(CAvideos.isnull().sum())

#Distribution of categories
print("\nDistribution of Categories IDs:")
print(CAvideos['category_id'].value_counts())

#Correlation matrix
print("\nCorrelation Matrix:")
corr_matrix = CAvideos[['views','likes','dislikes','comment_count']].corr()
print(corr_matrix)
```


## We need to map from the respective json file which has the category name for each ids, which will be useful for the later part of the analysis.
#### I am using the json file for US videos since they have all the category names. After a quick analysis with first using CA_category_id.json, I realised that they have missing values for category_id = 29. Hence to reduce the number of steps, I am direclty using the chosen US_category_id.json file

```python
#Mapping the categpories from the json file to the current dataset to better understand the date
import json

# Read the JSON file containing category information
with open('/kaggle/input/youtube-new/US_category_id.json', 'r') as file:
    category_data = json.load(file)

# Parse the JSON data to extract category names and IDs
category_mapping = {}
for category in category_data['items']:
    category_id = int(category['id'])
    category_name = category['snippet']['title']
    category_mapping[category_id] = category_name

    
# Assuming df is your DataFrame containing the YouTube dataset
# Map category IDs to category names in the dataset
CAvideos['category'] = CAvideos['category_id'].map(category_mapping)
#moving the category next to the ids

column = CAvideos.pop('category')
CAvideos.insert(5,'category',column)

```

# Fixing Data Types

```python
#changing boolean types to categories
CAvideos['comments_disabled']= CAvideos['comments_disabled'].astype('category')
CAvideos['ratings_disabled']=CAvideos['ratings_disabled'].astype('category')
CAvideos['video_error_or_removed']=CAvideos['video_error_or_removed'].astype('category')


#changing trending date type to datetime format
CAvideos['trending_date']= pd.to_datetime(CAvideos.trending_date,format='%y.%d.%m')

#Changing publish time to date time format and seperating to publish time and publish date
CAvideos['publish_time'] =pd.to_datetime(CAvideos['publish_time'], errors='coerce', format = '%Y-%m-%dT%H:%M:%S.%fZ')

#checking if the changes have implemented
CAvideos.info()
CAvideos.head(3)
```

```python
#Analysing the categorical data
comments_disabled_counts = CAvideos['comments_disabled'].value_counts()
ratings_disabled_counts = CAvideos['ratings_disabled'].value_counts()
video_error_or_removed_counts = CAvideos['video_error_or_removed'].value_counts()
print("Comments Disabled Counts:\n", comments_disabled_counts)
print("Ratings Disabled Counts:\n", ratings_disabled_counts)
print("Video Error Counts or removed Counts:\n", video_error_or_removed_counts)
```

# Visualizing the Data


## Plotting the correlation to have a better understanding the relationship between views, likes, dislikes and comment count

### 1.Matrix Heatmap

```python
#plotting the stronger correlated variables
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='cividis', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### 2.Scatter plots to further analyse the strength

```python
plt.figure(figsize=(15, 10))
plt.subplot(2,3,1)
sns.scatterplot(x='views',y='likes', data=CAvideos, alpha=0.5)
plt.title("Views vs Likes")
plt.xlabel("Views")
plt.ylabel("Likes")

plt.subplot(2,3,2)
sns.scatterplot(x='views',y='dislikes', data=CAvideos, alpha=0.5)
plt.title("Views vs Dislikes")
plt.xlabel("Views")
plt.ylabel("Dislikes")

plt.subplot(2,3,3)
sns.scatterplot(x='views',y='comment_count', data=CAvideos, alpha=0.5)
plt.title("Views vs Comment Count")
plt.xlabel("Views")
plt.ylabel("Comment Count")


plt.subplot(2,3,4)
sns.scatterplot(x='likes',y='comment_count', data=CAvideos, alpha=0.5)
plt.title("Likes vs Comment Count")
plt.xlabel("Likes")
plt.ylabel("Comment Count")

plt.subplot(2,3,5)
sns.scatterplot(x='comment_count',y='dislikes', data=CAvideos, alpha=0.5)
plt.title("Comment Count vs Dislikes")
plt.xlabel("Comment Count")
plt.ylabel("Dislikes")


plt.tight_layout()
plt.show()

```

### 3.Analysing the categories with the highest likes-dislikes ratio with bar plot


```python
#printing the count of each categories
print(CAvideos.category.value_counts())
#Exploring ratio of likes-dislikes per categories
likesdf = CAvideos.groupby('category')['likes'].agg('sum')
# calculating total dislikes for each category
dislikesdf = CAvideos.groupby('category')['dislikes'].agg('sum')
# calculating ratios of likes to dislikes
ratiodf = likesdf/dislikesdf 
# most liked category to appear on top
ratiodf = ratiodf.sort_values(ascending=False).reset_index()
# plotting bar chart
ratiodf.columns = ['category','ratio']
plt.subplots(figsize=(10, 15))
sns.barplot(x="ratio", y="category", data=ratiodf,
            label="Likes-Dislikes Ratio", color="green")
```
